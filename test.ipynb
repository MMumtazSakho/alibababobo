```python
import PyPDF2

def extract_text_from_pdf(pdf_file):
    with open(pdf_file, 'rb') as file:
        reader = PyPDF2.PdfReader(file)
        text = ''
        for page in reader.pages:
            text += page.extract_text()
    return text

pdf_file = "animalia-8-16-2-4.pdf"
text = extract_text_from_pdf(pdf_file)
print(text)
```

```python
        
from http import HTTPStatus
import dashscope
import re
dashscope.base_http_api_url = 'https://dashscope-intl.aliyuncs.com/api/v1'
dashscope.api_key='sk-a1311b4902ae4d34bc944af0165b52e5'
def call_with_stream(materi):
    messages = [
        {'role': 'user', 
         'content': f"""
            Kamu adalah AI yang bertugas untuk membuat pertanyaan pilihan ganda dengan 3 opsi. 
            kalimat pertanyaan diawali dengan <q>, sedangkan opsi jawaban diawali dengan <o>, sedangkan kunci jawaban diawali dengan <a>.
            contoh:
            <q> berapa nilai dari 1+1 ?
            <o> A. 2
            <o> B. 1
            <o> C. 3
            <a> jawabannya adalah A. 2
            buatlah hanya 1 pertanyaan saja berdasarkan materi ini:
            {materi}          
        """}]
    
    responses = dashscope.Generation.call("qwen-max",
                                messages=messages,
                                result_format='message',  # set the result to be "message" format
                                stream=True,              # set streaming output
                                incremental_output=True   # get streaming output incrementally
                                )
    
    full_text = ""
    for response in responses:
        if response.status_code == HTTPStatus.OK:
            full_text += response.output.choices[0]['message']['content']
        else:
            error_message = ('Request id: %s, Status code: %s, error code: %s, error message: %s' % (
                response.request_id, response.status_code,
                response.code, response.message
            ))
            return error_message
    print(full_text)
    question = re.search(r'<q>(.*?)\s*(?=<o>)', full_text, re.DOTALL)
    options = re.findall(r'<o>\s*(.*?)\s*(?=<|$)', full_text, re.DOTALL)
    answer = re.search(r'<a>(.*?)$', full_text, re.DOTALL)
    result = {
        "question": question.group(1).strip() if question else None,
        "options": [opt.strip() for opt in options],
        "answer": answer.group(1).strip() if answer else None
    }
    
    return result
```
